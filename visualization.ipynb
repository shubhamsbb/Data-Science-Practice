```python
x = np.arange(1,201,2)
y = np.random.randint(1,10,100) 
```

```python
plt.figure(figsize=(15,5))
plt.axes(facecolor='pink')
plt.plot(x, y,color='black',label='Line chart') #Line Chart
plt.xlim(0, 200) 
# plt.xticks(1)
plt.ylim(0, 10)
plt.xlabel('Sales',color='blue')
plt.ylabel('product')
plt.title('Line Plot')
plt.suptitle('Sales Comparison', size=20, y=1.05)
plt.savefig('Save.png')
plt.legend()
plt.show()
```

```python
mtcars.hist(figsize=(15,5))
plt.show()
```

```python
# table
pd.crosstab(mtcars.gear,mtcars.cyl)
```

```python
mtcars.plot()
```

```python
# bar plot between 2 different categories
pd.crosstab(mtcars.gear,mtcars.cyl).plot(kind="bar")
plt.show()
```

```python
mtcars["gear"].value_counts()
mtcars.gear.value_counts().plot(kind="pie")
plt.show()
```

```python
plt.figure(facecolor='green')
plt.axes()
plt.scatter(mtcars.mpg,mtcars.qsec,marker='o')## scatter plot of two variables
plt.title('Scatterplot',color='blue')
plt.xlabel('mpg',color='red',size=10)
plt.ylabel('qsec',color='green',size=10)
plt.show()
```

# variables are correlated

```python
# graphical representation of data 
# import matplotlib.pyplot as plt
# histogram
plt.hist(mtcars["mpg"])
plt.show()
```

```python

# craetes histogram with 5bins and colours filled init
plt.hist(mtcars["mpg"],facecolor='peru',edgecolor="blue",bins=3)
plt.show()
```

```python
#Boxplot
#help(plt.boxplot)
plt.boxplot(mtcars["mpg"])
plt.show()
```

```python
# vertical Boxplot
plt.boxplot(mtcars['mpg'],vert =False)
plt.ylabel("MPG")
plt.xlabel("Boxplot")
plt.title("Boxplot")  
```

```python
#violin plot
#help(plt.violinplot)
plt.violinplot(mtcars["mpg"])
plt.show()
```
