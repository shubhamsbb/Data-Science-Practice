```python
#import library
import numpy as np
import pandas as pd

from sklearn.model_selection import train_test_split 
from sklearn.tree import DecisionTreeClassifier 
from sklearn.metrics import accuracy_score ,classification_report, confusion_matrix ,ConfusionMatrixDisplay
```

```python
#import data
fraud_data = pd.read_csv("https://raw.githubusercontent.com/slmsshk/DataSet/main/Fraud_check.csv")
```

```python
fraud_data
```

```python
#convert categorical value

bins = [0,30000, np.inf]
names = ['Risk', 'Good']

fraud_data['Taxable.Income'] = pd.cut(fraud_data['Taxable.Income'], bins, labels=names)

fraud_data['Taxable.Income']
```

```python
fraud_data
```

```python
# for convert categorical to int

categorical = [col for col in fraud_data.columns if fraud_data[col].dtype==object]

for col in categorical:
    temp = {}
    count = 0
    for val in fraud_data[col].values:
        try:
            temp[val]
        except:
            temp[val] = count
            count += 1
    fraud_data[col] = [temp[x] for x in fraud_data[col].values]
print(fraud_data[categorical])
```

```python
fraud_data
```

```python
#feture selection
y = fraud_data['Taxable.Income']
fraud_data.drop(['Taxable.Income'], axis=1, inplace=True)
X = fraud_data
del fraud_data
```

```python
# Split dataset into training set and test set
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=1) # 70% training and 30% test
```

```python
# Create Decision Tree classifer object
clf = DecisionTreeClassifier()

# Train Decision Tree Classifer
clf = clf.fit(X_train,y_train)

#Predict the response for test dataset
y_pred = clf.predict(X_test)
```

```python
# model Confusion Accuracy
    
print("Report : ",classification_report(y_test, y_pred) 
) 
```

```python
confusion_matrix(y_test, y_pred)
```

```python
accuracy_score(y_test,y_pred)*100
```

```python
ConfusionMatrixDisplay.from_predictions(y_test,y_pred)
```
